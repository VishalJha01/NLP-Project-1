```python
import nltk
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')   # <-- add this line
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
data = {
    'Question': [
        "How much is the admission fee?",
        "How can I apply for a hostel?",
        "When will exams start?",
        "Where can I check my timetable?",
        "What is the last date to pay fees?"
    ],
    'Answer': [
        "Admission fee is â‚¹5000.",
        "Fill the hostel form online at hostel.university.edu.",
        "Exams will begin in December as per the academic calendar.",
        "You can check your timetable on the university website under 'Academics'.",
        "The last date to pay fees is 30th September."
    ]
}
faq_df = pd.DataFrame(data)
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word.isalnum() and word not in stop_words]
    return ' '.join(tokens)

faq_df['Processed_Q'] = faq_df['Question'].apply(preprocess)
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(faq_df['Processed_Q'])
```

```python
def chatbot_response(user_input):
    user_input_processed = preprocess(user_input)
    user_vec = vectorizer.transform([user_input_processed])
    similarity = cosine_similarity(user_vec, tfidf_matrix)
    index = similarity.argmax()
    return faq_df.iloc[index]['Answer']
```

```python
print("ðŸŽ“ University FAQ Chatbot (type 'exit' to quit)")
while True:
    query = input("You: ")
    if query.lower() == 'exit':
        print("Bot: Goodbye! ðŸ‘‹")
        break
    response = chatbot_response(query)
    print("Bot:", response)
```
